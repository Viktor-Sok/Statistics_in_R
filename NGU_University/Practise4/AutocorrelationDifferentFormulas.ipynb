```r
#time series has periodic theoretiocal correlation function Kxx(t,t+tau) = cos(w*tau)
ts <- sin(seq(0, 8*pi, pi/100 ))
```

```r
str(ts)
```

```r
plot(ts)
```

# Pearson-like correlation (correspond to theoretical correlation coefficient much better) 

```r
lag <- seq(2, 800, 1) # difference in time between two realizations of the proccess
```

```r
str(lag)
```

```r
n <- length(ts)
```

```r
corr_values <- vector()
```

```r
for (var in lag) {
    corr_values <- append(corr_values,  cor( ts[1:(n - var)], ts[(var + 1):n] ))
}
```

```r
plot(corr_values, ylim=c(-1,1), cex = 0.5, type = 's')
```

# Default R formula

```r
acf(ts, lag.max = 500, type = "correlation", plot= TRUE)
```
