# Loading and Exploring Data

```r
data <- read.csv('dataRegression.csv', dec = '.', sep= ';')
```

```r
str(data)
```

```r
clicksOnTuesday <- data[data$DayOfWeek ==2,1]
```

```r
clicksOnWednesday <- data[data$DayOfWeek == 3,1]
```

```r
str(clicksOnTuesday)
```

```r
str(clicksOnWednesday)
```

```r
summary(clicksOnTuesday)
```

```r
summary(clicksOnWednesday)
```

# One Sample Test on a specific value of the mean in the population

```r
wilcox.test(clicksOnTuesday, mu = 11)
```

```r
wilcox.test(clicksOnWednesday, mu = 12)
```

```r
#converting LogNormal to Normal distribution
logClicksTuesday <- log(clicksOnTuesday)
logClicksWednesday <- log(clicksOnWednesday)
```

```r
summary(logClicksTuesday)
```

```r
boxplot(logClicksTuesday)
```

```r
t.test(logClicksTuesday, mu = 2)
```

# Two Sample Test on the equality of the means

```r
wilcox.test(clicksOnTuesday, clicksOnWednesday)
```

```r
t.test(logClicksTuesday, logClicksWednesday )
```

*From the tests one can conclude that the average number of clicks on Tuesday and Wednesday are the same*

# Two Sample Test on the equality of the standart deviations

```r
mood.test(clicksOnTuesday, clicksOnWednesday)
```

*Mood test says that the standart deviations are likely the same*
