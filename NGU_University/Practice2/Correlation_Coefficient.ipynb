```r
data <- read.csv("dataRegression.csv", sep =";", dec=".")
```

```r
str(data)
```

# Plotting correlation matix

```r
library("corrplot") # plots correleation matrix
```

```r

corrplot(cor(data[,1:6]),"ellipse", tl.cex = 0.1)
```

# Scatterplot Clicks/Calls 

```r
plot(data$Clicks, data$Calls)
```

# Calculating correlation coefficient

```r
#Pearson's corr. coef.
cor(data$Clicks, data$Calls,method = "pearson")
```

```r
# Checking statistical significance of pearson's corr. coef. using Student's t-test
cor.test(data$Clicks, data$Calls, alternative = "two.sided", method = "pearson")
```
